## Peak hour of sales

```python
# Extract the hour from the time
final_merged['hour'] = final_merged['time'].dt.hour
final_merged['hour']
```

```python
# Group by hour and sum the quantity sold
sales_by_hour = final_merged.groupby('hour')['quantity'].sum().sort_values(ascending=False)
```

```python
print(sales_by_hour) ## the peak sales oocur around 12pm and early afternoon 1pm with highest order count of 6776 and 6413 respectively 
```

### Day of the week is when sales are made the most

```python
# Get the day of the week (e.g., Monday, Tuesday, etc.)
final_merged['day_of_week'] = final_merged['date'].dt.day_name()
final_merged['day_of_week'] 
```

```python
# Calculate total sales (price × quantity) per day of the week
sales_by_day = final_merged.groupby('day_of_week').apply(
    lambda df: (df['price'] * df['quantity']).sum()
).sort_values(ascending=False)
```

```python
print("Total sales by day of the week:")
print(sales_by_day)
```

### Top 5 Bestselling Pizzas

```python
# Group by pizza name and sum quantities
top_pizzas = final_merged.groupby('name')['quantity'].sum().sort_values(ascending=False).head(5)

print("Top 5 Bestselling Pizzas:")
print(top_pizzas)
```

### Sales made in each Month

```python
# Create a 'month' column in a readable format (January, Febuary, March, etc)
final_merged['month'] = final_merged['date'].dt.to_period('M').astype(str)
final_merged['month'] 
```

```python
# Group by month and calculate total sales (price * quantity)
monthly_sales = final_merged.groupby('month').apply(
    lambda df: (df['price'] * df['quantity']).sum()
).sort_index()

print("Monthly Sales:")
print(monthly_sales)
```

### Pizza types that are not doing well on the menu

```python
pizza_sales = final_merged.groupby('name')['quantity'].sum().sort_values()
print("Least Selling Pizzas:")
print(pizza_sales.head(5)) 
```
